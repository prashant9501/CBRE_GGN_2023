# Paired Sample t-test

The data shows the weekly average alcohol (in ml) consumption before and after breakup. Conduct a  paired t-test to check whether the alcohol consumption is more after the breakup  (that is $\mu_d$ > 0) at 95% confidence ($\alpha$ = 0.05).  

```python
import pandas as pd
import numpy as np
from scipy.stats import norm, t
```

```python
df = pd.read_csv('breakups.csv')
df['Difference'] = df['After_Breakup'] - df['Before_Breakup']
df.head()
```

```python
df.shape
```

```python
# Assuming the differences follow a normal distribution
# D = mean diff in the estimated parameter value before and after the treatment (breakup or training prog)
# S_d = Standard Deviation
# mu_d = hypothesized mean difference

D = df['Difference'].mean()
S_d = df['Difference'].std()
mu_d = 0

n = len(df)  # smaple size

D, S_d
```

```python
# H0: mu_d <= 0
# HA: mu_d > 0    >>> RIGHT-TAILED TEST
    
# t_dist with (n-1) dof
alpha = 0.05
t_stat = (D-mu_d) / (S_d/np.sqrt(n))
t_stat
```

```python
t.isf(alpha, df=n-1)  # crit value
# Decion: FAIL TO REJECT H0
```

```python
t.sf(t_stat, df=n-1)  # p > alpha 
```

## Example 2:

A researcher believes that people drink more coffee on Mondays than other days of  the week. Based on a sample of 50 coffee drinkers, the mean difference was estimated  as 14 ml and the corresponding standard deviation was 8.5 ml. Conduct an appropriate hypothesis test at a = 0.1 to check the claim that people drink on average 10 ml  more coffee on Mondays compared to other days of the week.


```python
# HA: Diff > 10 ml   ### means Right-tailed test
# H0: Diff <= 10 ml   
    
n = 50   # sample size
alpha = 0.1   # significance level
sample_mean_difference = 14
sample_std = 8.5
```

```python
SE = sample_std/np.sqrt(n)
t_stat = (sample_mean_difference - 10)/SE
t_stat
```

```python
t.isf(alpha, df=n-1)  # crit value
```

```python
t.sf(t_stat, df=n-1)  # p << alpha 
```

```python
# Decision: REJECT H0
```

Claim that people drink on average 10 ml  more coffee on Mondays compared to other days of the week.... IS VALID !
